```python
n = int(input("Enter number of items: "))

wt = []
val = []

print("\nEnter weight and value for each item:")
for i in range(n):
    w = int(input(f"Weight of item {i+1}: "))
    v = int(input(f"Value of item {i+1}: "))
    wt.append(w)
    val.append(v)

W = int(input("\nEnter maximum capacity of knapsack: "))


def knapsack(W, wt, val, n):
    dp = [[0 for _ in range(W + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, W + 1):
            if wt[i - 1] <= w:
                dp[i][w] = max(val[i - 1] + dp[i - 1][w - wt[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][W]

print("Maximum value in knapsack =", knapsack(W, wt, val, n))
```
