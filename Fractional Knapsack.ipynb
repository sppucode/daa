```python
n = int(input("Enter number of items: "))

weights = []
values = []

print("\nEnter weight and value for each item:")
for i in range(n):
    w = int(input(f"Weight of item {i+1}: "))
    v = int(input(f"Value of item {i+1}: "))
    weights.append(w)
    values.append(v)

W = int(input("\nEnter maximum capacity of knapsack: "))
items = list(zip(weights, values))

# Sort items by value/weight ratio (high to low)
items.sort(key=lambda x: x[1]/x[0], reverse=True)
capacity = W

total_value = 0

for w, v in items:
    if capacity == 0:
        break
    if w <= capacity:
        total_value += v
        capacity -= w
    else:
        total_value += v * (capacity / w)
        capacity = 0

print(f"Maximum value = {total_value:.2f}")
```
